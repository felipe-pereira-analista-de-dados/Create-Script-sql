```python
import random
import pandas as pd

# Função para gerar datas aleatórias
def random_date():
    return random.choice(['2023-01-15', '15-02-2023', '2023/03/05', '2023-04-20', '2023-05-10', '2023-06-25'])

# Função para gerar emails aleatórios
def random_email():
    domains = ["example.com", "test.com", "fake.com"]
    return f"user{random.randint(1, 1000)}@{random.choice(domains)}"

# Gerar dados desorganizados para 'clientes'
clientes = []
for i in range(1, 100):
    nome = f"Cliente {random.randint(1, 500)}"
    email = random_email() if random.choice([True, False]) else None
    data_nascimento = random.choice(['1985-02-15', '03/08/1990', '12/25/1982', '1977-11-30', '1989/07/13', None])
    clientes.append((i, nome, email, data_nascimento))

# Gerar dados desorganizados para 'produtos'
produtos = []
for i in range(1, 100):
    nome = f"Produto {i}"
    preco = random.choice(['29.99', '49.99', '19.95', '99.90', None])
    produtos.append((i, nome, preco))

# Gerar dados desorganizados para 'vendas'
vendas = []
for i in range(1, 100):
    cliente_id = random.randint(1, 100)
    produto_id = random.randint(1, 100)
    quantidade = random.choice([1, 2, 3, None])
    data_venda = random_date()
    vendas.append((i, cliente_id, produto_id, quantidade, data_venda))

# Gerar dados desorganizados para 'pagamentos'
pagamentos = []
for i in range(1, 100):
    venda_id = i
    metodo_pagamento = random.choice(['Cartão de Crédito', 'Dinheiro', 'Cartão de Débito', 'Boleto', 'Pix'])
    valor_pagamento = round(random.uniform(10.0, 1000.0), 2)
    data_pagamento = random_date()
    pagamentos.append((i, venda_id, metodo_pagamento, valor_pagamento, data_pagamento))

# Criação do script SQL
sql_script = """
-- Criação das tabelas
CREATE TABLE clientes (
    cliente_id INT PRIMARY KEY,
    nome VARCHAR(255),
    email VARCHAR(255),
    data_nascimento VARCHAR(255)
);

CREATE TABLE produtos (
    produto_id INT PRIMARY KEY,
    nome VARCHAR(255),
    preco DECIMAL(10, 2)
);

CREATE TABLE vendas (
    venda_id INT PRIMARY KEY,
    cliente_id INT,
    produto_id INT,
    quantidade INT,
    data_venda VARCHAR(255),
    FOREIGN KEY (cliente_id) REFERENCES clientes(cliente_id),
    FOREIGN KEY (produto_id) REFERENCES produtos(produto_id)
);

CREATE TABLE pagamentos (
    pagamento_id INT PRIMARY KEY,
    venda_id INT,
    metodo_pagamento VARCHAR(255),
    valor_pagamento DECIMAL(10, 2),
    data_pagamento VARCHAR(255),
    FOREIGN KEY (venda_id) REFERENCES vendas(venda_id)
);

-- Inserção de dados desorganizados na tabela 'clientes'
INSERT INTO clientes (cliente_id, nome, email, data_nascimento) VALUES
"""

for cliente in clientes:
    sql_script += str(cliente) + ",\n"
sql_script = sql_script.rstrip(",\n") + ";\n"

sql_script += """
-- Inserção de dados desorganizados na tabela 'produtos'
INSERT INTO produtos (produto_id, nome, preco) VALUES
"""

for produto in produtos:
    sql_script += str(produto) + ",\n"
sql_script = sql_script.rstrip(",\n") + ";\n"

sql_script += """
-- Inserção de dados desorganizados na tabela 'vendas'
INSERT INTO vendas (venda_id, cliente_id, produto_id, quantidade, data_venda) VALUES
"""

for venda in vendas:
    sql_script += str(venda) + ",\n"
sql_script = sql_script.rstrip(",\n") + ";\n"

sql_script += """
-- Inserção de dados desorganizados na tabela 'pagamentos'
INSERT INTO pagamentos (pagamento_id, venda_id, metodo_pagamento, valor_pagamento, data_pagamento) VALUES
"""

for pagamento in pagamentos:
    sql_script += str(pagamento) + ",\n"
sql_script = sql_script.rstrip(",\n") + ";\n"



# Salvar o script em um arquivo .sql
file_path = 'C:\\Users\\Felipe\\Downloads\\TESTE\\loja_mysql.sql'
with open(file_path, 'w') as file:
    file.write(sql_script)

file_path
```
